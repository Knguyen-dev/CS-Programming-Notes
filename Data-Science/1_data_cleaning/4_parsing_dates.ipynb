# Parsing Dates
Here we'll work on parsing dates

```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
landslides = pd.read_csv("../data/catalog.csv")

# set our random seed for reproducibility
np.random.seed(0)
```

```python
'''
1. Check data type of our date column. So it looks like the dates 
are in the form 'mm/dd/yy'. Pandas uses type 'object' for storing various types 
of data, but most commonly you'd see this for strings.

'''
landslides.head()

'''
2. We'll need to convert the values in this column to things that are 
programmatically recognizable as dates. The strings in the series
can parsed using the following matcher: '%m/%d/%y'. 

We'll create a new feature/column called 'date_parsed'. As a result 
the datatype for this column should be datetime64, which is good.
- NOTE 1: https://strftime.org/

- NOTE 2: If there are multiple date formats within one column, then you
can let pandas try to infer the right date format. Of course, pandas won't 
always be able to figure things out, and also it's much slower than specifying the 
exact format.

'''
landslides['date_parsed'] = pd.to_datetime(landslides['date'], format="%m/%d/%y")

# Inferring 
# landslides['date_parsed'] = pd.to_datetime(landslides['Date'], infer_datetime_format=True)

# Get a series of only the days
day_of_month_landslides = landslides['date_parsed'].dt.day

# Drop missing values
day_of_month_landslides = day_of_month_landslides.dropna()

sns.histplot(day_of_month_landslides, kde=False)
```
